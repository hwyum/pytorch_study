```
from data_load import movie_data
```

```
dataset = movie_data('ratings.txt')
```

```
data = dataset.data[['document', 'label']]
data = data[~data.document.isna()]
```

```
from konlpy.tag import Mecab
import gluonnlp as nlp
import itertools
from sklearn.model_selection import train_test_split
mecab = Mecab()
```

```
documents = list(data.document)
labels = list(data.label)

# tokenize
data_tokenized = [(mecab.morphs(d),l) for d,l in zip(documents, labels)]

# train / test split
tr_data, tst_data = train_test_split(data_tokenized , test_size=0.2)
```

```
tr_data[1]
```

```
# build vocab --> tr data 만 활용함
counter = nlp.data.count_tokens(itertools.chain.from_iterable([d for d,l in tr_data]))
vocab = nlp.Vocab(counter=counter, min_freq=10)
```

```
# connecting embedding to vocab
ptr_embedding = nlp.embedding.create('fasttext',source='wiki.ko')
vocab.set_embedding(ptr_embedding)
```

```
import pickle
# saving vocab
with open('../data/vocab.pkl', mode = 'wb') as io:
    pickle.dump(vocab, io)
```
